```python
# Função para detectar outliers em uma coluna do conjunto de dados
def verificar_outliers(coluna):
    q1 = np.quantile(coluna, .25)
    q3 = np.quantile(coluna, .75)
    iqr = q3 - q1

    limite_inferior = q1 - (1.5 * iqr)
    limite_superior = q3 + (1.5 * iqr)

    outliers = [x for x in coluna if x < limite_inferior or limite_superior < x]

    total_outliers = len(outliers)
    freq_relativa_outliers = round(total_outliers/coluna.shape[0],3)*100

    print(f'Coluna: "{coluna.name}"\nTotal : {coluna.shape[0]}\nLimite Inferior: {limite_inferior}\nLimite Superior: {limite_superior}\nOutliers: {total_outliers}\nFrequência: {freq_relativa_outliers}%')

    
# Função para criar mapa de calor do Coeficiente de Correlação de Pearson em uma coluna numérica
def mapa_de_correlacao(dados):
    plt.figure(figsize = (7,7))
    sns.heatmap(dados.corr("spearman"), annot = True, cmap = "YlGnBu")
    plt.title(f"Mapa de Correlação - {(dados.tipo_imovel.unique()[0])} \n", fontsize = 15)
    plt.show();


# Função para criar gráfico de Barras Horizontal sobre Frequência de Tipo de Imóvel em Bairros
def barh_bairros_tipo_imovel(dados, tamanho):
    # dados -> conjunto de dados
    # tamanho -> número de labels no gráfico
    
    fig, ax = plt.subplots(figsize = (8,6))
    
    x = list(dados['bairro'].value_counts()[:tamanho].index)
    y = list(dados['bairro'].value_counts()[:tamanho])

    # definindo paleta de cores padrão do gráfico    
    ax = sns.barplot(x = y, y = x, width = 0.5, palette = sns.blend_palette(['#0D3B66','#C2EABD'], tamanho))
 
    # informando valor correspondente de cada barra
    ax.bar_label(ax.containers[0], fontsize=10);

    # removendo eixo X
    plt.gca().axes.get_xaxis().set_visible(False)
    
   
    # condicional para definir título
    if dados.tipo_imovel.nunique() == 1:
        ax.set_title(f'Bairros com mais {str(dados.tipo_imovel.unique()[0])+"s"}', fontsize = 15, fontfamily = 'Sans', fontweight = 'bold')
    else:
        ax.set_title(f'Bairros com mais Tipos de Imóveis', fontsize = 15, fontfamily = 'Sans', fontweight = 'bold')
    
    # removendo bordas
    ax.spines[['top','left','right','bottom']].set_visible(False)
    
    # ajustando layout
    fig.tight_layout();
    
    
     # condicional para salvar arquivo .png
    if int(dados.tipo_imovel.nunique()) == 1:
        plt.savefig(f'pics/bairros_{dados.tipo_imovel.unique()[0]}s')
    else:
         plt.savefig(f'pics/bairros_com_mais_tipos_imoveis')

    
# Função para criar gráfico de Barras com Total de Imóveis 
def barv_total_imoveis():

    # criando figura com subplot
    fig, ax = plt.subplots(figsize = (6,4))

    # calculando os valores de X e Y
    x = list(aluguel['tipo_imovel'].value_counts().index)
    y = list(aluguel['tipo_imovel'].value_counts())

    # definindo ordem de cor das barras do gráfico
    cores = ['#0D3B66','#FAF0CA','#FAF0CA', '#FAF0CA']

    # criando gráfico de Barras
    ax= sns.barplot(x = x, y = y, width = 0.5, palette = cores)

    # inserindo título
    ax.set_title('Total de Imóveis por Tipo', fontsize = 15, fontfamily = 'Sans', fontweight = 'bold')

    # removendo valores do eixo Y
    plt.gca().axes.get_yaxis().set_visible(False)

    # inserindo os valores equivalentes para cada barra
    ax.bar_label(ax.containers[0], fontsize=10);

    # removendo bordas
    ax.spines[['top','left','right','bottom']].set_visible(False)

    # ajustando layout
    fig.tight_layout();

    # salvando gráfico em formato .png
    plt.savefig(f'pics/total_tipos_imoveis')


# Função para criar Boxplot do Valor de Aluguel por Tipo de Imóvel  
def boxplot_aluguel_tipo_imovel():

    boxplot_valor_aluguel= sns.boxplot(y = aluguel.valor_aluguel,
                                        x = aluguel.tipo_imovel,
                                       width = 0.3,
                                       orient = 'v',
                                       color = 'deepskyblue')
    # removendo nomes dos eixos
    boxplot_valor_aluguel.set_ylabel('')
    boxplot_valor_aluguel.set_xlabel('')

    # definindo tamanho da figura
    boxplot_valor_aluguel.figure.set_size_inches(8,6)

    # definindo título
    boxplot_valor_aluguel.set_title('Valor de Aluguel por Tipo de Imóvel');
    
    
# Função para criar Histograma do Valor de Aluguel por Tipo de Imóvel      
def histograma_valor_aluguel_tipo_imovel():

    # criando subplot
    fig, axes = plt.subplots(2,2, figsize = (14,8))

    # criando Histograma de Valor de Aluguel por Tipo de Imóvel
    sns.histplot(ax = axes[0,0], data = apartamentos, x = 'valor_aluguel', color = 'darkorange', bins = 50)

    sns.histplot(ax = axes[0,1], data = kitnet_studio.query('valor_aluguel <= 15000'), x = 'valor_aluguel', color = 'dodgerblue', bins = 50)

    sns.histplot(ax = axes[1,0], data = casas, x = 'valor_aluguel', color =  'red', bins = 50)

    sns.histplot(ax = axes[1,1], data = casas_condominio, x = 'valor_aluguel', color ='limegreen', bins = 50)

    # inserindo títulos
    axes[0,0].set_title('Valor do Aluguel de Apartamentos')
    axes[0,1].set_title('Valor do Aluguel de Kitnet Studio')
    axes[1,0].set_title('Valor do Aluguel de Casas')
    axes[1,1].set_title('Valor do Aluguel de Casas em Condomínio')

    # renomeando eixos 
    axes[0,0].set_xlabel('Valor de Aluguel')
    axes[0,1].set_xlabel('Valor de Aluguel')
    axes[1,0].set_xlabel('Valor de Aluguel')
    axes[1,1].set_xlabel('Valor de Aluguel')


    # ajusatando layout
    fig.tight_layout();    


# Função para criar Boxplot da Área por Tipo de Imóvel      
def boxplot_area_tipo_imovel():
    boxplot_area = sns.boxplot( x = aluguel.area,
                               y = aluguel.tipo_imovel,
                               width = 0.3,
                               orient = 'h',
                               color = 'limegreen')

    boxplot_area.set_ylabel('')
    boxplot_area.set_xlabel('')

    boxplot_area.figure.set_size_inches(8,6)
    boxplot_area.set_title('Área em Metros por Tipo de Imóvel');
    

# Função para criar Histograma da Área por Tipo de Imóvel  
def histograma_area_tipo_imovel():
    # criando subplot
    fig, axes = plt.subplots(2,2, figsize = (14,8))

    # criando Histograma de Área em m2 por Tipo de Imóvel
    sns.histplot(ax = axes[0,0], data = apartamentos, x = 'area', color = 'darkorange', bins = 50)

    sns.histplot(ax = axes[0,1], data = kitnet_studio, x = 'area', color = 'dodgerblue', bins = 50)

    sns.histplot(ax = axes[1,0], data = casas, x = 'area', color =  'red', bins = 50)

    sns.histplot(ax = axes[1,1], data = casas_condominio, x = 'area', color ='limegreen', bins = 50)

    # inserindo títulos
    axes[0,0].set_title('Área em Apartamentos')
    axes[0,1].set_title('Área em Kitnet Studio')
    axes[1,0].set_title('Área em Casas')
    axes[1,1].set_title('Área em Casas em Condomínio')

    # renomeando eixos 
    axes[0,0].set_xlabel('Área')
    axes[0,1].set_xlabel('Área')
    axes[1,0].set_xlabel('Área')
    axes[1,1].set_xlabel('Área')


    # ajusatando layout
    fig.tight_layout();

# Função para criar Boxplot do Aluguel do Imóvel de acordo com Quantidade de Quartos
def boxplot_aluguel_qtd_quartos(imovel):
    boxplot_aluguel_quartos = sns.boxplot(imovel, x = 'valor_aluguel', y = 'quartos', orient = 'h', width = 0.35, color = 'orange')

    # Ajustando layout do gráfico
    boxplot_aluguel_quartos.set_title(f'Boxplot - Valor de Aluguel por Qtd de Quartos em {imovel.tipo_imovel[0]}s')
    boxplot_aluguel_quartos.figure.set_size_inches(9,6)
    boxplot_aluguel_quartos.set_xlabel('Preço (R$)')
    boxplot_aluguel_quartos.set_ylabel('Qtd de Quartos')

    boxplot_aluguel_quartos.figure.tight_layout()
    
# Função para criar Boxplot do Aluguel do Imóvel de acordo com Quantidade de Quartos
def boxplot_aluguel_qtd_garagens(imovel):
    boxplot_aluguel_quartos = sns.boxplot(imovel, x = 'valor_aluguel', y = 'garagem', orient = 'h', width = 0.4, color = 'yellow')

    # Ajustando layout do gráfico
    boxplot_aluguel_quartos.set_title(f'Boxplot - Valor de Aluguel por Qtd de Garagens em {imovel.tipo_imovel[0]}s')
    boxplot_aluguel_quartos.figure.set_size_inches(9,6)
    boxplot_aluguel_quartos.set_xlabel('Preço (R$)')
    boxplot_aluguel_quartos.set_ylabel('Qtd de Garagens')

    boxplot_aluguel_quartos.figure.tight_layout()
    
# Função para criar Gráfico de Dispersão entre Valor de Aluguel e Área por Tipo de Imóvel
def dispersao_area_aluguel_tipo_imovel():

    # criando subplot
    fig, axes = plt.subplots(2,2, figsize = (12,8))

    # criando gráfico de Dispersão entre Área e Valor de Aluguel por Tipo de Imóvel
    sns.scatterplot(ax = axes[0,0], data = apartamentos, x = 'area', y = 'valor_aluguel', c = 'darkorange')

    sns.scatterplot(ax = axes[0,1], data = kitnet_studio, x = 'area', y = 'valor_aluguel', c = 'dodgerblue')

    sns.scatterplot(ax = axes[1,0], data = casas, x = 'area', y = 'valor_aluguel', c =  'red')

    sns.scatterplot(ax = axes[1,1], data = casas_condominio, x = 'area', y = 'valor_aluguel', c ='limegreen')

    # inserindo títulos
    axes[0,0].set_title('Dispersão em Apartamentos')
    axes[0,1].set_title('Dispersão em Kitnet Studio')
    axes[1,0].set_title('Dispersão em Casas')
    axes[1,1].set_title('Dispersão em Casas em Condomínio')

    # renomeando eixos 
    axes[0,0].set_ylabel('Valor de Aluguel')
    axes[0,0].set_xlabel('Área')

    axes[0,1].set_ylabel('Valor de Aluguel')
    axes[0,1].set_xlabel('Área')

    axes[1,0].set_ylabel('Valor de Aluguel')
    axes[1,0].set_xlabel('Área')

    axes[1,1].set_ylabel('Valor de Aluguel')
    axes[1,1].set_xlabel('Área')

    # ajusatando layout
    fig.tight_layout();
    

# Função para criar Gráfico em Pares para avaliar a distribuição e a associação das variáveis     
def pairplot_aluguel():
    sns.pairplot(aluguel, hue = 'tipo_imovel', kind = 'reg', corner = True, markers = ["o", "s", "D", "^"], height = 3)
    
def pairplot_por_imovel():
    for imovel in aluguel.tipo_imovel.unique():   
        pairplot_imovel = sns.pairplot(aluguel.query('tipo_imovel in @imovel'), kind = 'reg', y_vars='valor_aluguel' ,x_vars=['area', 'quartos','garagem'], height = 4, plot_kws = {'line_kws':{'color':'red'}})
    
        # Alternando labels do Eixo Y
        label_y = ['Preço (R$)']
        for ax, lbl in zip(pairplot_imovel.axes.flatten(), label_y):
            ax.set_ylabel(f'{lbl}')
            
        # Alternando labels do Eixo X            
        label_x = ['Área','Quartos', 'Garagens']
        for ax, lbl in zip(pairplot_imovel.axes.flatten(), label_x):
            ax.set_xlabel(f'{lbl}')
        
        # Ajustando Títulos e layout
        pairplot_imovel.figure.suptitle(f'Dispersão e Reta de Regressão - {imovel}s', fontweight = 'bold')
        pairplot_imovel.figure.tight_layout()
    
```

# Valor do Aluguel

## Teste de Normalidade - Valor do Aluguel

* O Teste de Normalidade consiste em avaliar estatisticamente se uma variável possui um Distribuição Normal ou não.

* Será adotado um **Nível de Confiança** de **95%** para todos os Testes de Normalidade utilizados neste projeto.

* O Critério adotado para validar a Aceitação ou Rejeição da Hipótese Nula ($H_0$) é o $p$$-value$

### Importando Teste de Normalidade 

```python
from scipy.stats import normaltest
```

### Definindo Nível de Significância

```python
significancia_normaltest = 0.05
```

### Formulação de Hipóteses

$H_0$: O Valor de Aluguel para cada Tipo de Imóvel segue uma Distribuição Normal

$H_1$: O Valor de Aluguel para cada Tipo de Imóvel não segue uma Distribuição Normal

### Critério $p$$-value$

> ### Rejeitar $H_0$ se $p$$-value \leq \alpha$

```python
# Teste de Normalidade para Valor de Aluguel em Apartamentos
_, p_value_aluguel_apartamentos = normaltest(aluguel.query('tipo_imovel == "Apartamento"').valor_aluguel)

p_value_aluguel_apartamentos <= significancia_normaltest
```

```python
# Teste de Normalidade para Valor de Aluguel em Studio e Kitnet
_, p_value_aluguel_kitnet = normaltest(aluguel.query('tipo_imovel == "Studio e kitnet"').valor_aluguel)

p_value_aluguel_kitnet <= significancia_normaltest
```

```python
# Teste de Normalidade para Valor de Aluguel em Casas
_, p_value_aluguel_casa = normaltest(aluguel.query('tipo_imovel == "Casa"').valor_aluguel)

p_value_aluguel_casa <= significancia_normaltest
```

```python
# Teste de Normalidade para Valor de Aluguel em Casas em Condomínio
_, p_value_aluguel_casa_condominio = normaltest(aluguel.query('tipo_imovel == "Casa em condomínio"').valor_aluguel)

p_value_aluguel_casa_condominio <= significancia_normaltest
```

### Resultado - Teste de Normalidade

* A Hipótese Nula de que o Valor de Aluguel segue uma Distribuição Normal em cada Tipo de Imóvel está rejeitada com 95% de Confiança. Portanto, o Valor de Aluguel para cada Tipo de Imóvel não possui uma Distribuição Normal e revela que há possíveis Assimetrias na Distribuição destes dados.

## Teste de Kruskal-Wallis - Comparação de Valor de Aluguel por Tipo de Imóvel

* Como não há uma Distribuição Normal do Valor de Aluguel e os Tipos de Imóveis são Independentes, será utilizado o Teste de Kruskal-Wallis para compreender se há uma difença significativa entre os Valores do Aluguel para cada Tipo de Imóvel

* Será assumido um Nível de Significância ($\alpha$) de 5% para validar a Aceitação ou Rejeição da Hipótese Nula 

* O Critério adotado para Aceitação ou Rejeição da $H_0$ é o $p$$-value$

### Formulando Hipótese 

###  $H_0$: *Não há uma diferença significativa para o Preço em cada Tipo de Imóvel*

### $H_1$: *Dois Tipos de Imóveis ou mais possuem uma diferença significativa*

### Definindo Nível de Significância ($\alpha$)

```python
significancia_kruskal = 0.05
```

### Critério $p$$-value$

> ### Rejeitar $H_0$ se $p$$-value \leq \alpha$

```python
from scipy.stats import kruskal
```

```python
_, p_value_kruskal = kruskal(apartamentos['valor_aluguel'], kitnet_studio['valor_aluguel'], casas['valor_aluguel'], casas_condominio['valor_aluguel'])
```

```python
p_value_kruskal <= significancia_kruskal
```

## Resultado - Teste de Kruskal Wallis 

* A $H_0$ de que as Medianas das Valores do Aluguel por Tipo de Imóvel são iguais está rejeitada com 95% de Confiança. Portanto, há uma diferença estatisticamente significativa para o Valor de Aluguel para cada Tipo de Imóvel

## Pós-Teste de Dunn's

* Após realizar o Teste de Kruskal-Wallis, se a $H_0$ foi Rejeitada, pode-se aplicar o Pós-Teste de Dunn's para uma comparação múltiplas entre as categorias e e validar quais possuem uma diferença significativa.

* O Nível de Confiança adotado é de **95%**

* Será adotado o $p-$$value$ para validar a $H_0$

### Formulando Hipóteses

$H_0$: Casa e Casa em Condomínio não possuem uma diferença significativa em relação ao Valor de Aluguel

$H_1$: Casa e Casa em Condomínio comparados apresentam uma diferença significativa em relação ao Valor de Aluguel

****

$H_0$: Studio e Kitnet não possue diferença significativa do Valor de Aluguel comparado à cada Tipo de Imóvel

$H_1$: Studio e Kitnet possui uma diferença significativa do Valor de Aluguel comparado à cada Tipo de Imóvel

```python
import scikit_posthocs as sp
```

```python
p_value_teste_dunn = sp.posthoc_dunn(aluguel, 'valor_aluguel', 'tipo_imovel', 'bonferroni')
```

### Critério $p$$-value$

> ### Rejeitar $H_0$ se $p$$-value \leq \alpha$

```python
p_value_teste_dunn <= 0.05
```

## Resultado - Pós-Teste de Dunn's

* A $H_0$ para o Valor de Aluguel entre Casas e Casas em Condomínio está aceita com 95% de Confiança, portanto, não há uma diferença significativa entre Casas e Casas em Condomínio quanto ao Valor do Aluguel

* A $H_0$ sobre Studio e Kitnet está rejeitada, portanto, ao compará-la com cada um dos outros Tipos de Imóveis, há sim uma diferença significativa para o Valor do Aluguel

# Modelagem Preditiva

## Modelos Preditivos utilizados

No momento, os modelos que serão utilizados com este conjunto de dados são os seguintes:

* **Regressão Linear Simples**
* **Regressão Linear Múltipla**


## Variáveis

As variáveis preditoras selecionadas foram: **Área**, **Quantidade  de Quartos** e **Quantidade de Vagas/Garagens**, isto porque estas apresentaram uma boa correlação com o variável predita **Valor de Aluguel**, mas não uma correlação muito forte e que afetaria o desempenho do modelo preditivo, sendo medidas utilizando o **Coeficiente de Correlação de Pearson**.

* Para a **Regressão Linear Simples** será utilizada a variável **Área** como variável preditora (`feature`) e a variável **Valor de Aluguel** como variável predita (`target`)

* Para a Regressão Linear Múltipla serão utilizadas as variáveis **Área**,**Quartos** e **Garagem** como variáveis preditoras (`features`) e a variável **Valor de Aluguel** como variável predita (`target`)

**OBSERVAÇÃO**: As **features** selecionadas apresentaram uma **correlação fraca** com o **target**, portanto, **não será criado** um modelo preditivo do Valor de Aluguel para **Studio e Kitnet**

## Funções para Regressão Simples e Múltipla

Com intuito de otimizar código e economizar tempo, além de proporcionar mais flexibilidade e facilidade de uso, foram criadas duas funções para cada modelo citado acima. As funções podem receber os seguintes parâmetros:

   * **df** = DataFrame completo dos dados do Tipo de Imóvel que serão utilizados para Modelagem Preditiva

```python
# importando dependências
from pandas import DataFrame
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error, median_absolute_error

# Função para modelo de Regressão Linear Simples 
def reg_linear_simples_aluguel(df):
    
    # Selecionando Feature e Target
    dados = df[['area','valor_aluguel']]
    
    X = dados.iloc[:,:1]
    y = dados.iloc[:,-1]
    
    # Separando os conjuntos de Treino e Teste
    X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.3, random_state = 101)
    
    # Instanciando o Modelo
    regressao = LinearRegression()
    
    # Estimando modelo com os conjuntos de Treino
    regressao.fit(X_treino, y_treino)
    
    # Criando Previsões com o dados de Teste da feature
    y_pred = regressao.predict(X_teste)
    
    # Armazenando dados de Teste e Predições para trasnformar em dataframe
    np.set_printoptions(precision = 2)
    comparacao_y = np.concatenate((y_pred.reshape(len(y_pred),1), np.array(y_teste).reshape(len(y_teste),1)),1)
    
    # Obtendo Equação da Reta e Métricas de Avaliação
    print(f'Equação da Reta: {regressao.intercept_:.2f} + {float(regressao.coef_):.2f}x\n\nCoeficiente de Determinação (Treino): {regressao.score(X_treino,y_treino).round(3)}\nCoeficiente de Determinação (Teste): {r2_score(y_teste,y_pred).round(3)}\n\nErro Médio Absoluto(MAE): {mean_absolute_error(y_teste, y_pred).round(3)}\nErro Mediano Absoluto (MedAE): {median_absolute_error(y_teste,y_pred).round(3)} \nRaíz do Erro Quadrático Médio (RMSE): {np.sqrt(mean_squared_error(y_teste, y_pred)).round(3)}\n\n')
    
    # Criando figura para plotagem
    fig, axs = plt.subplots(1,3, figsize = (17, 5))
    
    # Gráfico com Reta da Regressão estimada
    axs[0].scatter(X_treino, y_treino)
    axs[0].plot(X_teste, regressao.predict(X_teste), color = 'red')
    axs[0].set_title(f'Regressão Linear Simples - Aluguel de {df.tipo_imovel.unique()[0]}')
    axs[0].set_xlabel('Área')
    axs[0].set_ylabel('Preço')
    
    # Calculando Resíduos
    residuos = y_teste - y_pred
    
    # Plotando Distribuição dos Resíduos para avaliar Normalidade
    axs[1].hist(residuos)
    axs[1].set_title(f'Distribuição dos Resíduos')
    axs[1].set_xlabel('Resíduos')
    
    # Dispersão entre Valores Previstos e Resíduos para avaliar Homocedasticidade
    axs[2].scatter(x = y_pred, y = residuos**2)
    axs[2].set_title(f'Dispersão - Previsão x Resíduos²')
    axs[2].set_xlabel('Resíduos')
    
    # Salvando Previsões e Valor Real
    resultado = DataFrame(comparacao_y, columns = ['Previsão','Real'])
    resultado.to_csv(f'Regressão Simples - {df.tipo_imovel.unique()[0]}s.csv', index = False)
    
    
    
    
    

# Função para modelo de Regressão Linear Múltipla
def reg_linear_multipla_aluguel(df):
      
    dados = df[['area','quartos','garagem','valor_aluguel']]
    
    X = dados.iloc[:,:3]
    y = dados.iloc[:,-1]
    
    X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.3, random_state = 101)
    
    regressao = LinearRegression(normalize = False)
    
    regressao.fit(X_treino, y_treino)
    
    y_pred = regressao.predict(X_teste)
    
    np.set_printoptions(precision = 2)

    comparacao_y = np.concatenate((y_pred.reshape(len(y_pred),1),np.array(y_teste).reshape(len(y_teste),1)),1)
    
    r2_ajustado = (1 - ((1 - r2_score(y_teste, y_pred)) * (len(y_teste) - 1)) / 
          (len(y_teste) - X_treino.shape[1] - 1))
    
    print(f'Equação da Reta: ({float(regressao.coef_[0]):.2f}x1) + ({float(regressao.coef_[1]):.2f}x2) + ({float(regressao.coef_[2]):.2f}x3) + ({regressao.intercept_:.2f})\n\nCoeficiente de Determinação (Treino): {regressao.score(X_treino,y_treino).round(3)}\nCoeficiente de Determinação (Teste): {r2_score(y_teste,y_pred).round(3)}\n\nErro Médio Absoluto(MAE): {mean_absolute_error(y_teste, y_pred).round(3)}\nErro Mediano Absoluto (MedAE): {median_absolute_error(y_teste,y_pred).round(3)} \nRaíz do Erro Quadrático Médio (RMSE): {np.sqrt(mean_squared_error(y_teste, y_pred)).round(3)}\n\n')

    
    # Calculando Resíduos
    residuos = y_teste - y_pred

    # Criando figura para plotagem
    fig, axs = plt.subplots(1,2, figsize = (15, 5))
    
    # Plotando Distribuição dos Resíduos
    axs[0].hist(residuos)
    axs[0].set_title(f'Distribuição dos Resíduos')
    axs[0].set_xlabel('Resíduos')
    
    # Dispersão entre Valores Previstos e Resíduos
    axs[1].scatter(x = y_pred, y = residuos**2)
    axs[1].set_title(f'Dispersão - Previsão x Resíduos²')
    axs[1].set_xlabel('Resíduos')
    
    
    resultado = DataFrame(comparacao_y, columns = ['Previsão','Real'])
    
    resultado.to_csv(f'Regressão Múltipla - {df.tipo_imovel.unique()[0]}s.csv', index = False)
```

## Métricas de Avaliação

A seguir são definidas as métricas que serão utilizadas para avaliação dos modelos de Regressão . Como trata-se de preços de diferentes imóveis e na Análise Exploratória notou-se a presença de muitos outliers em todos os Tipos de Imóveis, foram adotas as métricas da Raíz do Erro Quadrático Médio para compreender como está a performance do modelo quantos aos outliers e do Erro Mediano Absoluto para medir qual o erro na estimava de preço em 50% dos valores preditos.

* **Coeficiente de Determinação (R2)**: Medir o quanto o modelo consegue explicar a variância em valor percentual (%)
* **Raíz do Erro Quadrático Médio (RMSE)**: Medida sensível à valores discrepantes para penalizar erros grandes admitindo a Mediana como minimizador
* **Erro Mediano Absoluto (MedAE)**: Mede o erro absoluto até o percentil 50% (Mediana)
